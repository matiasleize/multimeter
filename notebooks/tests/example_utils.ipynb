```python
from utils import *  # noqa: F403,F401


def N_sigma_to_L_and_N_eff(N_sigma,Nd):
    PTE = n_sigma_to_PTE(N_sigma)
    L = PTE_to_L(PTE=PTE, Nd=Nd)
    N_eff = L_iso_to_n_eff(d=2, L=L)
    return L, N_eff

```
